```python
import numpy as np
import pandas as pd
```

```python
data = {"Hoten": ['nva', 'nvb'],
        "Toan": [7.5, 8],
        "Ly": [6, 9],
        "Hoa": [6.5, 5.5]}
df = pd.DataFrame(data)
df
```

```python
df_new1 = pd.melt(frame=df, id_vars='Hoten', value_vars=['Toan', 'Ly', 'Hoa'])
df_new1
```

```python
df_new2 = pd.melt(frame=df, id_vars='Hoten', value_vars=['Toan', 'Ly', 'Hoa'], var_name='Monthi', value_name='Diem')
df_new2
```

```python
df_new3 = pd.melt(frame=df, id_vars='Hoten', var_name='Monthi', value_name='Diem')
df_new3
```

### Pivoting Data

```python
df_pivot = df_new3.pivot(index='Hoten', columns='Monthi', values='Diem')
df_pivot
```

```python
# thêm 1 dòng vào df_new3
df_new4 = df_new3.append({"Hoten": 'nvb', 'Monthi': 'Hoa', 'Diem': 7}, ignore_index=True)
df_new4
```

```python
df_pivot2 = df_new4.pivot(index='Hoten', columns='Monthi', values='Diem')
df_pivot2
```

```python
df_pivot2 = df_new4.pivot_table(index='Hoten', columns='Monthi', values='Diem', aggfunc=np.mean)
df_pivot2
```
