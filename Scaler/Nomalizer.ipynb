```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

## Normalizer

```python
# định cỡ lại để độ dài (khoảng cách từ các điểm đến trung tâm đều có giá trị là 1)
# giả sử có 3 tính năng là x, y z ==> công thức của Normalizer
# xi / sqrt(xi2 + yi2 + zi2)
```

```python
df = pd.DataFrame({
    'A': np.random.randint(-100, 100, 1000).astype(float),
    'B': np.random.randint(-80, 80, 1000).astype(float),
    'C': np.random.randint(-150, 150, 1000).astype(float)
})
df.head()
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
f = plt.figure(figsize=(6,6))
ax = f.add_subplot(111, projection='3d')
ax.scatter(df['A'], df['B'], df['C'])
plt.show()
```

```python
from sklearn.preprocessing import Normalizer
```

```python
scaler = Normalizer()
```

```python
df_scale = scaler.fit_transform(df)
df_scale = pd.DataFrame(df_scale, columns=df.columns)
df_scale.head()
```

```python
f = plt.figure(figsize=(6,6))
ax = f.add_subplot(111, projection='3d')
ax.scatter(df_scale['A'], df_scale['B'], df_scale['C'])
plt.show()
```
