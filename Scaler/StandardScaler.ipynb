```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
np.random.seed(1)
df = pd.DataFrame({
    'A': np.random.normal(0, 2, 10000),
    'B': np.random.normal(5, 3, 10000),
    'C': np.random.normal(-5, 5, 10000)
})
df.head()
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1,3,1)
sns.distplot(df['A'])
plt.subplot(1,3,2)
sns.distplot(df['B'])
plt.subplot(1,3,3)
sns.distplot(df['C'])
plt.show()
```

```python
df['A'].skew(), df['B'].skew(), df['C'].skew()
```

```python
f, ax1 = plt.subplots(ncols=1, nrows=1, figsize=(6,5))
sns.distplot(df['A'], ax = ax1, hist=False)
sns.distplot(df['B'], ax = ax1, hist=False)
sns.distplot(df['C'], ax = ax1, hist=False)
plt.show()
```

```python
f, ax1 = plt.subplots(ncols=1, nrows=1, figsize=(6,5))
sns.kdeplot(df['A'], ax = ax1)
sns.kdeplot(df['B'], ax = ax1)
sns.kdeplot(df['C'], ax = ax1)
plt.show()
```

```python
# using StandardScaler
from sklearn.preprocessing import StandardScaler
```

```python
#khởi tạo StandarScaler
scaler = StandardScaler()
```

```python
df_scale = scaler.fit_transform(df)
df_scale = pd.DataFrame(df_scale, columns=df.columns)
df_scale.head()
```

```python
f, ax1 = plt.subplots(ncols=1, nrows=1, figsize=(6,5))
sns.kdeplot(df_scale['A'], ax = ax1)
sns.kdeplot(df_scale['B'], ax = ax1)
sns.kdeplot(df_scale['C'], ax = ax1)
plt.show()
```

```python
f, (ax1, ax2) = plt.subplots(ncols=2, nrows=1, figsize=(6,5))
sns.kdeplot(df['A'], ax = ax1)
sns.kdeplot(df['B'], ax = ax1)
sns.kdeplot(df['C'], ax = ax1)
sns.kdeplot(df_scale['A'], ax = ax2)
sns.kdeplot(df_scale['B'], ax = ax2)
sns.kdeplot(df_scale['C'], ax = ax2)
plt.show()
```

```python
print('Before scale:', df['A'].mean(), df['B'].mean(), df['C'].mean())
print('After scale:', df_scale['A'].mean(), df_scale['B'].mean(), df_scale['C'].mean())
```

```python
#độ lệch chuẩn
print('Before scale:', df['A'].std(), df['B'].std(), df['C'].std())
print('After scale:', df_scale['A'].std(), df_scale['B'].std(), df_scale['C'].std())
```
